## Daily Death Cases in India

```python
import plotly.graph_objects as go
```

```python
# Mortality rate
mor = corona_in_india.groupby(['Date'])['Deaths','Total cases','Active cases'].sum().reset_index().sort_values('Date',
                                                                                                              ascending = True )
mor['Mortality Rate'] = (mor['Deaths'])/(mor['Total cases'])*100

fig = go.Figure()

fig.add_trace(go.Scatter(x = mor['Date'], y = mor['Mortality Rate'], mode = 'lines+markers', marker_color='red' ))

fig.update_layout(title = 'Mortality Rate in India')
fig.show(renderer='svg')
```

```python
data = corona_in_india.groupby(['Date'])['Deaths'].sum().reset_index().sort_values('Deaths',ascending = True)
data['Daily_Deaths'] = data['Deaths'].sub(data['Deaths'].shift())
data['Daily_Deaths'].iloc[0] = data['Deaths'].iloc[0]
data['Daily_Deaths'] = data['Daily_Deaths'].astype(int)

#Draw Figure
fig = px.bar(data , x = 'Date', y = 'Daily_Deaths', color = 'Daily_Deaths',height=500 ,title = 'Datewise Deaths in India')
fig.show(renderer='svg')
```

# Forecasting

```python
from fbprophet import Prophet
```

```python
corona_in_india.head()
```

```python
df = corona_in_india.groupby(['Date'])['Total cases'].sum().reset_index()
df.columns = ['ds','y']

m= Prophet()
m.fit(df)

future = m.make_future_dataframe(periods = 30)
forecast = m.predict(future)

fig = go.Figure()
fig.add_trace(go.Scatter(x = forecast['ds'], y = forecast['yhat'],
                        mode = 'lines+markers', name = 'Predict Cases'))

fig.add_trace(go.Scatter(x =  df['ds'], y = df['y'],
                        mode = 'lines+markers', name = 'Actual Cases'))

fig.update_layout(title = 'Forecasting of Total cases in India next 30 days')
fig.show(renderer='svg')
```

```python
df = corona_in_india.groupby(['Date'])['Deaths'].sum().reset_index()
df.columns = ['ds','y']

m= Prophet()
m.fit(df)

future = m.make_future_dataframe(periods = 30)
forecast = m.predict(future)

fig = go.Figure()
fig.add_trace(go.Scatter(x = forecast['ds'], y = forecast['yhat'],
                        mode = 'lines+markers', name = 'Predict Cases'))

fig.add_trace(go.Scatter(x =  df['ds'], y = df['y'],
                        mode = 'lines+markers', name = 'Actual Cases'))

fig.update_layout(title = 'Forecasting of Deaths in India next 30 days')
fig.show(renderer='svg')
```
